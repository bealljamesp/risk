### 2 - Core Risk Classifications

This taxonomy is the foundation for the entire FRM curriculum. You must know these boundaries perfectly.

```python
Market Risk

    Definition: The risk of financial loss due to movements in market prices.

    Sub-types: Interest rate risk, equity price risk, foreign exchange (FX) risk, and commodity price risk.
```

```python
Credit Risk

    Definition: The risk of economic loss stemming from a counterparty's failure to fulfill their contractual obligations.

    Sub-types: Default risk, bankruptcy risk, downgrade/migration risk, and counterparty credit risk (CCR) in OTC derivatives.

The Structural Identities: You must memorize these three components:

    Probability of Default (PD): The likelihood of a default event over a horizon.

    Loss Given Default (LGD): The percentage of exposure lost if default occurs (LGD = 1 - Recovery Rate).

    Exposure at Default (EAD): The gross dollar amount at risk at the moment of default.

The Expected Loss Equation:

    EL = PD * LGD * EAD

    ⚠️ Critical Quant Warning: GARP introduces EL here as a simple multiplication. In production, EAD for derivatives fluctuates stochastically over time based on underlying market factors, and PD and LGD are often correlated (wrong-way risk). Do not mistake this scalar equation for a simple static calculation in the real world.
```

```python
Operational Risk

    Definition: Loss resulting from inadequate or failed internal processes, people, systems, or external events. (Includes cyber risk and legal risk, but explicitly excludes strategic and reputational risk).

    RQ Significance: Hardest to model because data is sparse and fat-tailed. RQs use Extreme Value Theory EVT and loss distribution approaches LDA here.
```

```python
Liquidity Risk

    Funding Liquidity Risk: Inability to settle obligations immediately when they fall due without incurring unacceptable losses.

    Market Liquidity Risk: Inability to execute a transaction quickly at market price due to market asset thinness (reflected in wide bid-ask spreads).
```

### 4 - EL v UL

This is the most critical conceptual dichotomy in the entire chapter.

```python
Frequency
  ^
  |      [EL]     [ Unexpected Loss (UL) ]
  |     (Mean)   |<------- VaR -------->|
  |       |                             |
  |     ..*..                           |
  |    .  |  .                          |
  |   .   |   .                         |
  |  .    |    .                        |
  | .      |     .......................|......
  +---------------------------------------------> Loss
```

```python
Expected Loss (EL)

Definition: The average loss an institution anticipates over a given horizon.

Financial Management: EL is treated as a cost of doing business. It is covered directly by provisions and built into product pricing (e.g., credit card interest rates).
```

```python
Unexpected Loss (UL)

Definition: The volatility or variance of losses around the expected mean; the component of risk that cannot be forecasted precisely.

Financial Management: UL is covered by Economic Capital.

The Test Concept: Value at Risk (VaR) maps directly to UL at a specified confidence level. Capital reserves must be scaled to absorb UL to ensure structural solvency.
```

### 5 - Standard Risk Metrics

While these are developed fully in later chapters, know their axiomatic definitions now:

```python
VaR:

Purpose: Quantifies the maximum loss over a specified horizon at a given confidence level (1-alpha).

Weakness: Non-coherent: Fails subadditivity under specific non-normal conditions (the risk of a combined portfolio can appear larger than the sum of its parts).
```

```python
ES:

Purpose: Quantifies the EL conditional on theloss exceeding the VaR threshold.

Weakness: More computationally expensive; requires mapping th shape of the entire tail distribution.
```
