### 01-05 Modern Portfolio Theory (MPT) and the Capital Asset Pricing Model (CAPM)

$$\text{CML Equation}$$

$$R_i = R_F + \frac{\overline{R}_M - R_F}{\sigma_m^2} \sigma_{iM} = R_F + \frac{\overline{R}_M - R_F}{\sigma_M} \frac{\sigma_{iM}}{\sigma_M}$$

$R_i$ is the expected return of security $i$

$R_F$ is the risk-free rate

$\overline{R}_M$ is the expected return of the market portfolio

$\sigma_M$ is the standard deviation of the market portfolio

$\sigma_i$ is the standard deviation of security $i$

This equation shows that the expected return of a security is determined by its total risk (standard deviation) and the market risk premium per unit of total risk (Sharpe ratio of the market portfolio)

$\overline{R}_M - R_F$ is the market risk premium, which is the additional return that investors require for taking on the risk of investing in the market portfolio instead of the risk-free asset)

$\frac{\overline{R}_M - R_F}{\sigma_M}$ is the Sharpe ratio of the market portfolio, which measures the excess return per unit of risk for the market portfolio

*Well diversified portfolio does not have any idiosyncratic risk, so its expected return is determined solely by its systematic risk (beta) and the market risk premium*

Therefore, a well diversified portfolio will lie on the CML, which represents the risk-return tradeoff for efficient portfolios that can be formed by combining the risk-free asset and the market portfolio. On the other hand, individual securities and portfolios that are not fully diversified will have some idiosyncratic risk, which is uncorrelated with the market and can be diversified away in a well-diversified portfolio. Therefore, these securities and portfolios will lie on the SML, which represents the relationship between the expected return of an individual security and its systematic risk (beta).

$$\beta_i = \frac{\sigma_{iM}}{\sigma_m^2} = \rho_{iM}\frac{\sigma_{i}}{\sigma_m}$$

$\frac{\sigma_{i}}{\sigma_m}$ measures how a security's risk affects the market portfolio risk

$\sigma_{iM} = \rho_{iM} \sigma_i \sigma_M$

$\beta_i$ = 1 means the security has the same systematic risk as the market portfolio

$\beta_i$  > 1 means the security is more volatile than the market portfolio (higher systematic risk)

$\beta_i$  < 1 means the security is less volatile than the market portfolio (lower systematic risk)

$\beta_i$  = 0 means the security has no systematic risk (risk-free asset)

$\beta_{iM}$ is the portfolio's exposure to market's excess return

$$ \beta_{iM} = \rho_{i,M}\frac{\sigma_i}{\sigma_M} $$

or

$$ \beta_{(i,M)} = \frac{\sum_{t=1}^{n} (R_{it} - \overline{R}_i)(R_{Mt} - \overline{R}_M)}{\sum_{t=1}^{n} (R_{Mt} - \overline{R}_M)^2} $$

or

$$ \beta = \frac{\langle i,M \rangle}{\|M\|^2} $$



$\beta_{iM} <> \beta_{Mi}$ 

$\rho_{iM} = \rho_{Mi}$

The Two Types of NormalizationType A: Normalizing by Both Vector Lengths (Yields Correlation)If you normalize the dot product by dividing it by the length of the asset vector and the length of the market vector, you strip away all physical scale from both sides. This yields Correlation ($\rho$):$$ \rho = \frac{\langle \mathbf{i}, \mathbf{M} \rangle}{\|\mathbf{i}\| \|\mathbf{M}\|} = \cos(\theta) $$The Result: A pure, unitless measure of directional alignment locked strictly between $-1$ and $+1$. It tells you how closely the two vectors trace the same path, completely ignoring their actual size.Type B: Normalizing by the Base Vector Length Squared (Yields Beta)If you normalize the dot product by dividing it strictly by the squared length of the Benchmark Vector ($\|\mathbf{M}\|^2$), you only strip away the scale of the market. This yields Beta ($\beta$):$$ \beta = \frac{\langle \mathbf{i}, \mathbf{M} \rangle}{\|\mathbf{M}\|^2} $$The Result: A scale-dependent sensitivity coefficient. It preserves the raw physical size of the asset ($\|\mathbf{i}\|$), but frames it as a ratio of the market's size. It answers the question: "For every 1 unit of length the market vector extends, how many units along that same direction does the asset vector extend?"

$\text{Sharpe Ratio} = \frac{R_M - R_F}{\sigma_M}$
    - Measures the excess return per unit of risk for a portfolio

$\text{Treynor Ratio} = \frac{E(R_p) - R_f}{\beta_p}$
    - Measures the excess return per unit of systematic risk for a portfolio

$\text{Jensen's Alpha} = E(R_p) - [R_f + \beta_p * (E(R_m) - R_f)]$
    - Measures the excess return of a portfolio above the expected return predicted by the CAPM, which is based on the portfolio's beta and the market risk premium

$\text{Tracking Error} = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (R_{p,i} - R_{b,i})^2}$
    - Measures the standard deviation of the difference between the portfolio's returns and the benchmark's returns over a specified period

$\text{Information Ratio} = \frac{E(R_p) - E(R_b)}{\text{Tracking Error}}$
    - Measures the excess return of a portfolio above the benchmark per unit of tracking error

$\text{Sortino Ratio} = \frac{E(R_p) - R_f}{\sigma_d}$
    - Measures the excess return per unit of downside risk for a portfolio, where $\sigma_d$ is the standard deviation of negative returns (returns below a specified threshold, often the risk-free rate)
