```python
import pandas as pd

me_m = pd.read_csv(
    "../labs/data/portfolios_formed_on_me_monthly_ew.csv",
    header=0,
    index_col=0,
    parse_dates=True,
    na_values=-99.99,
)
```

```python
me_m.head()
```

```python
returns = me_m[["Lo 10", "Hi 10"]]
returns.columns = ["SmallCap", "LargeCap"]
returns = returns / 100
returns.head()
```

```python
returns.plot.line()
```

```python
returns.head()
```

```python
returns.index = pd.to_datetime(returns.index, format="%Y%m")
returns.head()
```

```python
returns.index = returns.index.to_period("M")
returns.head()
```
