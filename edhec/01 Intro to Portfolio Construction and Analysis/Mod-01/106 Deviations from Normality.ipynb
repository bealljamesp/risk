Skewness

$$\gamma_1 = E\left[\left(\frac{X - \mu}{\sigma}\right)^3\right] = \frac{E[(X - \mu)^3]}{\sigma^3} = \frac{\mu_3}{\sigma_3}$$


```python
erk.skewness(hfi)
```

```python
import scipy.stats

scipy.stats.skew(hfi)
```

```python
import numpy as np

normal_rets = np.random.normal(0, 0.15, size=(263, 1))
erk.skewness(normal_rets)
```

### Kurtosis

$$\beta_2 = E\left[\left(\frac{X - \mu}{\sigma}\right)^4\right] = \frac{E[(X - \mu)^4]}{\sigma^4} = \frac{\mu_4}{\sigma_4}$$


```python
erk.kurtosis(normal_rets)
```

```python
erk.kurtosis(hfi)
```

```python
scipy.stats.kurtosis(normal_rets, fisher=False)
```

```python
scipy.stats.jarque_bera(normal_rets)
```

```python
scipy.stats.jarque_bera(hfi)
```

```python
erk.is_normal(normal_rets)
```

```python
erk.is_normal(hfi)
```

```python
hfi.aggregate(erk.is_normal)
```
